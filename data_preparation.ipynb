```python
"""
Data Preparation for Unsupervised Learning
FIRST to prepare the data for all clustering algorithms
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("DATA PREPARATION FOR UNSUPERVISED LEARNING")
print("="*60)

# ============================================
# STEP 1: Load Dataset
# ============================================
print("\n STEP 1: Loading Dataset")
print("-"*40)

df = pd.read_csv('heart.csv')
print(f" Dataset loaded successfully!")
print(f"   Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"   Columns: {list(df.columns)}")

# ============================================
# STEP 2: Check for Missing Values
# ============================================
print("\n STEP 2: Checking for Missing Values")
print("-"*40)

missing = df.isnull().sum()
print(f"Missing values:\n{missing}")
print(f"Total missing: {missing.sum()}")

# ============================================
# STEP 3: Prepare Features (DROP Target for Unsupervised Learning)
# ============================================
print("\n STEP 3: Preparing Features for Clustering")
print("-"*40)

# Keep target only for validation, NOT for clustering
X = df.drop('target', axis=1)  # Features only - NO LABELS!
y_true = df['target']  # Keep for validation only

print(f" Features shape: {X.shape}")
print(f"   Features: {list(X.columns)}")
print(f"\n IMPORTANT: Target column DROPPED for unsupervised learning")
print(f"   (We won't use y_true for clustering - only for validation)")

# ============================================
# STEP 4: Scale Features (Important for distance-based algorithms)
# ============================================
print("\n STEP 4: Scaling Features")
print("-"*40)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f" Features scaled successfully!")
print(f"   Mean: {X_scaled.mean().mean():.2f}")
print(f"   Std: {X_scaled.std().mean():.2f}")

# ============================================
# STEP 5: Save Scaled Data
# ============================================
print("\n STEP 5: Saving Scaled Data")
print("-"*40)

X_scaled.to_csv('X_scaled.csv', index=False)
print(f" Scaled data saved to 'X_scaled.csv'")

# ============================================
# STEP 6: Quick Statistics
# ============================================
print("\n STEP 6: Data Statistics")
print("-"*40)

print(f"\n Original Data (before scaling):")
print(df.describe().round(2))

print(f"\n Disease Distribution:")
disease_rate = df['target'].mean() * 100
print(f"   Disease: {disease_rate:.1f}%")
print(f"   No Disease: {100-disease_rate:.1f}%")

# ============================================
# STEP 7: Save original data reference
# ============================================
print("\n STEP 7: Saving Data Info")
print("-"*40)

with open('data_info.txt', 'w') as f:
    f.write(f"Dataset: Heart Disease\n")
    f.write(f"Samples: {len(df)}\n")
    f.write(f"Features: {list(X.columns)}\n")
    f.write(f"Features count: {len(X.columns)}\n")
    f.write(f"Disease rate: {disease_rate:.1f}%\n")

print(f" Data info saved to 'data_info.txt'")

print("\n" + "="*60)
print(" DATA PREPARATION COMPLETE!")
print("="*60)
print("\n Next steps:")
print("   1. task1_kmeans.ipynb")
print("   2. task2_hierarchical.ipynb")
print("   3. task3_dbscan.ipynb")
print("   4. clustering_comparison.ipynb")
```
